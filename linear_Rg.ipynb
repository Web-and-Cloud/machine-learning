```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Data
X = np.array([1,2,3,4,5]).reshape(-1,1) #independent variables
y = np.array([2,4,5,4,5]) #dependent variables

#fit the linear regression model
model = LinearRegression()
model.fit(X,y)

#predicted line
y_pred = model.predict(X)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")

# Plot
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", label="Regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

n = len(X)
sum_X = np.sum(X)
sum_y = np.sum(y)
sum_X2 = np.sum(X**2)
sum_Xy = np.sum(X * y)

print(f"Sum X: {sum_X}, Sum y: {sum_y}")
print(f"Sum X^2: {sum_X2}, Sum Xy: {sum_Xy}")

# Slope (m)
m = (n * sum_Xy - sum_X * sum_y) / (n * sum_X2 - sum_X**2)
# Intercept (b)
b = (sum_y - m * sum_X) / n

print(f"Manual Slope (m): {m}")
print(f"Manual Intercept (b): {b}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Sample data
X1 = np.array([1, 2, 3, 4, 5])  # Independent variable 1
X2 = np.array([2, 4, 6, 8, 10])  # Independent variable 2
Y = np.array([50, 55, 65, 70, 75])  # Dependent variable

# Combine X1 and X2 into a 2D array
X = np.column_stack((X1, X2))

# Fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Coefficients and intercept
slope_X1 = model.coef_[0]
slope_X2 = model.coef_[1]
intercept = model.intercept_

print(f"Slope for X1: {slope_X1}")
print(f"Slope for X2: {slope_X2}")
print(f"Intercept: {intercept}")

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points
ax.scatter(X1, X2, Y, color="blue", label="Data Points")

# Create a meshgrid for the regression plane
X1_grid, X2_grid = np.meshgrid(
    np.linspace(min(X1), max(X1), 10),
    np.linspace(min(X2), max(X2), 10)
)
Y_grid = slope_X1 * X1_grid + slope_X2 * X2_grid + intercept

# Plot the regression plane
ax.plot_surface(X1_grid, X2_grid, Y_grid, color="yellow", alpha=0.5, label="Regression Plane")

# Label axes
ax.set_xlabel("X1")
ax.set_ylabel("X2")
ax.set_zlabel("Y")
ax.set_title("3D Linear Regression")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Data
X1 = np.array([1500, 2000, 2500, 3000, 3500])  # Size (sq ft)
X2 = np.array([3, 4, 4, 5, 5])  # Bedrooms
Y = np.array([250, 300, 350, 400, 450])  # Price ($1000s)

# Combine X1 and X2 into a 2D array
X = np.column_stack((X1, X2))

# Fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Coefficients and intercept
m1 = model.coef_[0]  # Coefficient for Size
m2 = model.coef_[1]  # Coefficient for Bedrooms
b = model.intercept_  # Intercept

print(f"Slope for Size (m1): {m1}")
print(f"Slope for Bedrooms (m2): {m2}")
print(f"Intercept (b): {b}")

# Predict the price for a house (2800 sq ft, 4 bedrooms)
house_features = np.array([[2800, 4]])
predicted_price = model.predict(house_features)
print(f"Predicted price for a 2800 sq ft house with 4 bedrooms: {predicted_price[0]} ($1000s)")

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot original data points
ax.scatter(X1, X2, Y, color="blue", label="Data Points")

# Create a grid for the regression plane
X1_grid, X2_grid = np.meshgrid(
    np.linspace(min(X1), max(X1), 10),
    np.linspace(min(X2), max(X2), 10)
)
Y_grid = m1 * X1_grid + m2 * X2_grid + b

# Plot the regression plane
ax.plot_surface(X1_grid, X2_grid, Y_grid, color="yellow", alpha=0.5, label="Regression Plane")

# Label axes
ax.set_xlabel("Size (sq ft)")
ax.set_ylabel("Bedrooms")
ax.set_zlabel("Price ($1000s)")
ax.set_title("3D Linear Regression")
plt.legend()
plt.show()
```
